```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
S=Symbol('S')
t=Symbol('t')
r=0.00163
sigma=0.01732
rho=0.01
R=0.5*(sigma**2)*(S**2)
```

```python
u0=S*sqrt(S)-(9.75*sqrt(S))
u0ss=diff(u0,S,S)
u0
```

```python
A0=u0ss**2
A0
```

```python
u1=-(-1*integrate(R*u0ss,t)-(r*S*diff(u0,S)))-(rho*(sigma**2)*-1*integrate((S**3)*A0,t))
u1
```

```python
u1ss=diff(u1,S,S)
A1=2*u0ss*u1ss
A1
```

```python
u2=-(-1*integrate(R*u1ss,t)-(r*S*diff(u1,S)))-(rho*(sigma**2)*-1*integrate((S**3)*A1,t))
u2
```

```python
u2ss=diff(u2,S,S)
A2=(2*u0ss*u2ss)+(u1ss**2)
A2
```

```python
u3=-(-1*integrate(R*u2ss,t)-(r*S*diff(u2,S)))-(rho*(sigma**2)*-1*integrate((S**3)*A2,t))
u3
```

```python
u3ss=diff(u3,S,S)
A3=(2*u0ss*u3ss)+(2*u1ss*u2ss)
A3
```

```python
u4=-(-1*integrate(R*u3ss,t)-(r*S*diff(u3,S)))-(rho*(sigma**2)*-1*integrate((S**3)*A3,t))
u4
```

```python
u4ss=diff(u4,S,S)
A4=(2*u0ss*u4ss)+(2*u1ss*u3ss)+(2*(u2ss**2))
A4
```

```python
u5=-(-1*integrate(R*u4ss,t)-(r*S*diff(u4,S)))-(rho*(sigma**2)*-1*integrate((S**3)*A4,t))
u5
```

```python
U=u0+u1+u2+u3+u4+u5
Uadm=U.subs({S:175.74,t:312})
Uadm
```
